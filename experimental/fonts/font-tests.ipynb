```python
#import pcface
import matplotlib.pyplot as plt
import numpy as np
import subprocess

def make_bits(c, m=1):
    #bitmap = pcface.make_bitmap(c, 'Px437_IBM_EGA_9x14.ttf', 16, 9, 14)
    #graph = pcface.make_graph(bitmap, 9, '01', False)
    #graph = graph.splitlines()
    graph = ["{:09b}".format(x) for x in FONT_MAP[c]]
    graph = [[int(x)*m for x in y] for y in graph]
    return graph

def make_text(t, rownum=0, tile=False, invtile=False):
    rows = None
    cn = 0
    for c in t:
        x = make_bits(c, m=255)
        if tile and ((cn % 2) ^ (rownum % 2)):
            x = [[z if z else 80 for z in y] for y in x]
        elif invtile and ((cn % 2) ^ (rownum % 2)):
            x = [[255 - z for z in y] for y in x]
                
        cn += 1
        if rows is None:
            rows = x
        else:
            for i in range(len(rows)):
                rows[i] += x[i]
            
    return rows

def make_texts(t, tile=False, invtile=False):
    arr = np.zeros((350, 720), dtype=np.uint8)
    i = 0
    wm = 0
    rn = 0
    
    lines = t.splitlines()
    maxlen = max(len(x) for x in lines)
    lines = [x.ljust(maxlen) for x in lines]
    
    for text in lines:
        line = make_text(text, rownum=rn+1, tile=tile, invtile=invtile)
        rn += 1
        
        h = len(line)
        w = len(line[0])
        arr[i:i+h, 0:w] = line
        i += h
        wm = max(w, wm)
        
    assert wm <= 720
    assert i <= 350
    return arr[:i, :wm]
```
